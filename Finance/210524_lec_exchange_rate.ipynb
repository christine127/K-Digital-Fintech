```python
%matplotlib inline
import matplotlib.pyplot as plt
import FinanceDataReader as fdr
import pandas_datareader as pdr
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings(action = 'ignore')
# https://fred.stlouisfed.org/series/GOLDAMGBD228NLBM 금값
# https://fred.stlouisfed.org/series/SLVPRUSD 은값
```

```python
def myreader(*args, src ='fred', sdate= '2000-01-01',edate=None, cview= True, col=None):
    df = pdr.DataReader(args, src, sdate)
    if col:
        df.columns = col
    print(df.shape)
    print(df.head())
    if cview:
        df.plot()
```

```python
myreader('GOLDAMGBD228NLBM','SLVPRUSD')
```
